```python
import pandas as pd 
```

```python
rosters = pd.read_csv('/Users/jordanwegner/Desktop/nba2/03_data/rosters.csv')
```

```python
teams = pd.read_csv('/Users/jordanwegner/Desktop/nba2/03_data/teams.csv')
df = rosters[['PLAYER','PLAYER_ID','TeamID','AGE']].merge(teams[['id','full_name']],how='left',left_on='TeamID',right_on='id')
df = df.sort_values(['PLAYER_ID','AGE'])
df.head()
```

```python
df['count'] = df.groupby('PLAYER_ID').cumcount()
```

```python
df = df.sort_values(['PLAYER_ID','count'])
```

```python
def HoopGrids(data,team1,team2):
    tdf = data[(data['full_name']==team1)|(data['full_name']==team2)][['PLAYER','full_name']].copy()
    tdf.drop_duplicates(inplace=True)
    data2 = pd.DataFrame(tdf['PLAYER'].value_counts().sort_values(ascending=False))
    data3 = data2[data2['PLAYER']>1].copy()
    return data3
```

```python
teams = sorted(df['full_name'].unique())
for i in teams:
    print(i)
```

```python
HoopGrids(df,'New York Knicks','Charlotte Hornets')
```

```python
HoopGrids(df,'New York Knicks','Chicago Bulls')
```
