```python
from nba_api.stats.endpoints import commonteamroster
ctr_list = []
for i in list(teams['id']):
    for j in list(seasons['id']):
        ctr_j = commonteamroster.CommonTeamRoster(team_id=i, season=j).get_data_frames()[0]
        ctr_list.append(ctr_j)
        time.sleep(.600) # had to do this to avoid the api timeout
        print("Team: {}, Season: {}".format(i,j))
```

```python
# okay... let's get all of the game ids... 
from nba_api.stats.endpoints import cumestatsplayergames
roster5_sample = roster5_plus.sample(5,random_state=41)

gids = []
for i in set(roster5_sample['id']):
    yr = roster5_sample[roster5_sample['id']==i]
    for j in yr['PLAYER_ID']:
        gid = cumestatsplayergames.CumeStatsPlayerGames(player_id=j,season=i).get_data_frames()[0]['GAME_ID'].values
        giddf = pd.DataFrame({'PLAYER_ID':j,'SEASON_ID':i,'GAME_IDS':gid})
        gids.append(giddf)
        
    print("Player ID: {}, Season: {}".format(i,j))
    
GIDS = pd.concat(gids)
GIDS
```

```python
gids_sample = GIDS.sample(10,random_state=41)
```

```python
# now, for the advanced box scores 
from nba_api.stats.endpoints import boxscoreadvancedv2

BS = []
for i in set(gids_sample['GAME_IDS']):
    bs = boxscoreadvancedv2.BoxScoreAdvancedV2(game_id=i).get_data_frames()[0]
    BS.append(bs)
    print(i)
BS_DF = pd.concat(BS)
```
